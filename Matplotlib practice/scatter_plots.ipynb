```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 90, 55, 60,65,  70, 85, 90]
```

```python
plt.scatter(study_hours, exam_scores)
plt.grid(True)
plt.xlabel("Study hours")
plt.ylabel("Exam scores")
plt.title("Study vs marks")
plt.show()
```

ADDING COLORS

```python
#colors=['red' if score < 50 else 'green' for score in exam_scores]
sizes=[score * 2 for score in exam_scores]
social_scores = [12, 34, 67, 555, 678, 145, 56, 999, 445]
plt.scatter(study_hours, exam_scores, c=social_scores, cmap='viridis')
plt.colorbar(label="Score")
plt.grid(True)
plt.xlabel("Study hours")
plt.ylabel("Exam scores")
plt.title("Study vs marks")
plt.show()
```

ADDING NNOTAIONS

```python
sizes=[score * 2 for score in exam_scores]
social_scores = [12, 34, 67, 555, 678, 145, 56, 999, 445]
plt.scatter(study_hours, exam_scores, c=social_scores, cmap='viridis')
plt.colorbar(label="Score")
plt.grid(True)

for i in range(len(study_hours)):
    plt.annotate(f'S {i+1}', (study_hours[i] +0.2, exam_scores[i]-2))
plt.xlabel("Study hours")
plt.ylabel("Exam scores")
plt.title("Study vs marks")
plt.show()
```

MULTIPLE GROUPS IN ONE PLOT

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]

class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label='Class A', color='Pink')
plt.scatter(class_b_hours, class_b_scores, label='Class B', color='Blue')

plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
