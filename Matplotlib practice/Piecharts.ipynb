```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot") #choose any
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
plt.show()
```

```python
colors =["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]
plt.pie(runs, labels=labels, colors = colors)
plt.show()
```

Adding Edges and Style Slices with wedgeprops

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor':'black', 'linewidth':2, 'linestyle':'--', 'fill': True}
)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor':'black', 'linewidth':2, 'linestyle':'--', 'fill': False}
)
plt.show()
```

HIGHLIGHT A SLICE USING EXPLODE

```python
explode = [0.1, 0, 0, 0]
plt.pie(runs, labels=labels, explode = explode, colors=colors)
plt.title("Explode Pie Example")
plt.show()
```

ADD SHADOWS FOR A 3D FEEL

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    explode=explode,
    shadow=True
)
plt.show()
```

START ANGLE

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    explode=explode,
    shadow=True,
    startangle=140
)
plt.show()
```

autopct FOR PERCENTAGE

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    explode=explode,
    shadow=True,
    startangle=140,
    autopct='%1.1f%%'
)
plt.show()
```
