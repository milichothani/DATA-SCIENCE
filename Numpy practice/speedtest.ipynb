```python
#Using python lists

import numpy as np
import time

size = 1_000_000

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x+y for x,y in zip(l1,l2)]
end = time.time()
print(add[0:10])
print(end - start)
```

```python
#Using NumPy arrays
import numpy as np
import time

size = 1_000_000

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
add = l1+l2
end = time.time()
print(add[0:10])
print(end - start)
```

```python
#Using python lists

import numpy as np
import time

size = 1_000_000

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x ** 2 for x in l1]
end = time.time()
print(add[0:10])
print(end - start)
```

```python
#Using NumPy arrays
import numpy as np
import time

size = 1_000_000

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
add = l1 ** 2
end = time.time()
print(add[0:10])
print(end - start)
```
